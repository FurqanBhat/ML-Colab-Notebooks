```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn
```

```python
X=torch.arange(-5,5,0.2).view(-1,1)
func=-5*X
```

```python
plt.plot(X.numpy(), func.numpy(), 'r',  label='func')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid('True', color='grey')
plt.show()
```

```python
Y=func + 1.7*torch.randn(X.size())
Y
```

```python
plt.plot(X.numpy(), Y.numpy(), 'b+',  label='Y')
plt.plot(X.numpy(), func.numpy(), 'r',  label='func')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid('True', color='grey')
plt.show()
```

```python
plt.plot(X.numpy(), Y.numpy(), 'b+',  label='Y')
```

```python
def forward(x): #prediction function, w=slope
  return w*x
```

```python
def criterion(y_pred, y): #loss function
  return torch.mean((y_pred-y)**2)
```

```python
w=torch.tensor(-10.0, requires_grad=True)
```

```python
step_size=0.1 #alpha
loss_list=[]
iter=20
```

```python
for i in range  (iter):
    Y_pred=forward(X)
    loss=criterion(Y_pred, Y)
    loss_list.append(loss.item())
    loss.backward()
    w.data=w.data-step_size*w.grad.data
    w.grad.data.zero_()
    print('{},\t{},\t{}'.format(i, loss.item(), w.item()))

```

```python
plt.plot(loss_list, 'r')
plt.tight_layout()
plt.grid('True', color='grey')
plt.xlabel('epoch/iterations')
plt.ylabel('loss')
plt.show()
```
